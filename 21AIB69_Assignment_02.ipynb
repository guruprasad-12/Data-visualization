```python
x=df.alcohol
y=df.quality
```

```python
plt.plot(x,y,color='yellow')
plt.title('PLOT')
plt.show()
```

```python
plt.stackplot(x,y,color='red')
plt.title('STACK PLOT')
plt.show()
```

```python
sns.barplot(x,y)
plt.title('BAR PLOT')
plt.show()
```

```python
sns.scatterplot(x,y)
plt.title('SCATTER PLOT')
plt.show()
```

```python
sns.histplot(x,color='green')
plt.title('HISTOGRAM PLOT')
plt.show()
```

```python
plt.stem(x,y)
plt.title('STEM PLOT')
plt.show()
```

```python
plt.step(x,y,data=None,c='orange')
plt.title('STEP PLOT')
plt.show()
```

```python
plt.pie(x,y,shadow=True)
plt.title('PIE PLOT')
plt.show()
```

```python
plt.polar(x,color='red')
plt.polar(y,color='yellow')
plt.title('POLAR PLOT')
plt.show()
```

```python
sns.countplot(x)
plt.title('COUNT PLOT')
plt.show()
```

```python
plt.boxplot(x)
plt.boxplot(y)
plt.title('BOX PLOT')
plt.show()
```

```python
sns.ecdfplot(x)
sns.ecdfplot(y)
plt.title('ECDF PLOT')
plt.show()
```

```python
sns.kdeplot(x,y,shadow=True,color='grey')
plt.title('KDE PLOT')
plt.show()
```

```python
df.plot.area()
plt.title('AREA PLOT')
plt.show()
```

```python
sns.lineplot(x,y)
plt.title('LINE PLOT')
plt.show()
```

```python
sns.heatmap(df.corr())
plt.title('HEATMAP PLOT')
plt.show()
```

```python
sns.distplot(x)
plt.title('DIST PLOT')
plt.show()
```

```python
sns.swarmplot(x)
plt.title('SWARM PLOT')
plt.show()
```
